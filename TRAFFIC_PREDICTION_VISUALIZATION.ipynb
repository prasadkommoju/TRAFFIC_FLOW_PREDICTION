```python
#.....GRAPH 1 _ Bar graph .................

import pandas as pd
import matplotlib.pyplot as plt

# Load the dataset
data = pd.read_csv("Traffic.csv")

# Group by day of week and sum the counts
grouped_data = data.groupby(['Day of the week']).sum()

# Plot
plt.figure(figsize=(10, 6))
bar_width = 0.2
index = range(len(grouped_data.index))

plt.bar(index, grouped_data['CarCount'], bar_width, label='CarCount')
plt.bar([i + bar_width for i in index], grouped_data['BusCount'], bar_width, label='BusCount')
plt.bar([i + 2*bar_width for i in index], grouped_data['TruckCount'], bar_width, label='TruckCount')
plt.bar([i + 3*bar_width for i in index], grouped_data['BikeCount'], bar_width, label='BikeCount')

plt.xlabel('Day of Week')
plt.ylabel('Total Count')
plt.title('Traffic Counts by Vehicle Type and Day of Week')
plt.xticks([i + 1.5*bar_width for i in index], grouped_data.index)
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the dataset
data = pd.read_csv("Traffic.csv")

# Plot histograms for TruckCount, BusCount, CarCount, and BikeCount
plt.figure(figsize=(12, 8))

plt.subplot(2, 2, 1)
plt.hist(data['TruckCount'], bins=20, color='blue', alpha=0.7)
plt.title('TruckCount Distribution')
plt.xlabel('TruckCount')
plt.ylabel('Frequency')

plt.subplot(2, 2, 2)
plt.hist(data['BusCount'], bins=20, color='green', alpha=0.7)
plt.title('BusCount Distribution')
plt.xlabel('BusCount')
plt.ylabel('Frequency')

plt.subplot(2, 2, 3)
plt.hist(data['CarCount'], bins=20, color='red', alpha=0.7)
plt.title('CarCount Distribution')
plt.xlabel('CarCount')
plt.ylabel('Frequency')

plt.subplot(2, 2, 4)
plt.hist(data['BikeCount'], bins=15, color='orange', alpha=0.7)
plt.title('BikeCount Distribution')
plt.xlabel('BikeCount')
plt.ylabel('Frequency')

plt.tight_layout()
plt.show()
```

```python
data=pd.read_csv("Traffic.csv")
df = pd.DataFrame(data)

# Assigning numerical values to Traffic Situation for plotting
traffic_situation_mapping = {'low':'low', 'normal': 'normal','heavy': 'heavy','high': 'high'}
df['Traffic Situation'] = df['Traffic Situation'].map(traffic_situation_mapping)

# Scatter plot
plt.figure(figsize=(10, 6))
plt.scatter(df['CarCount'], df['Traffic Situation'], color='red', label='CarCount')
plt.scatter(df['BusCount'], df['Traffic Situation'], color='blue', label='BusCount')
plt.scatter(df['TruckCount'], df['Traffic Situation'], color='green', label='TruckCount')
plt.scatter(df['BikeCount'], df['Traffic Situation'], color='orange', label='BikeCount')

plt.xlabel('Vehicle Count')
plt.ylabel('Traffic Situation')
plt.title('Vehicle Count vs Traffic Situation')
plt.legend()
plt.grid(True)
plt.show()
```

```python
import seaborn as sns
custom_dict = {"Monday": 0, "Tuesday": 1, "Wednesday":2, "Thursday": 3, "Friday": 4, "Saturday": 5, "Sunday": 6}

i = 0
for val in df['Traffic Situation'].unique():
    if val == "low":
        custom_dict[val] = 1
    if val == "normal":
        custom_dict[val] = 2
    if val == "high":
        custom_dict[val] = 3
    if val == "heavy":
        custom_dict[val] = 4

sns.catplot(data=df.copy().sort_values("Traffic Situation", key=lambda x: x.map(custom_dict)), x="Traffic Situation", y="Total", kind="bar")

plt.title("Average No.of Total Vehicles for different Traffic Situations")
plt.tight_layout()
```

```python
df["Hour"] = pd.to_datetime(df["Time"]).dt.hour

# Group the data by hour and sum the counts
hourly_traffic = df.groupby("Hour")[["CarCount", "BikeCount", "BusCount", "TruckCount"]].sum()

# Plotting
plt.figure(figsize=(10,6))
hourly_traffic.plot(kind="line", marker="o")
plt.xlabel("Hour of the day")
plt.ylabel("Total Count")
plt.title("Traffic by Hour of the Day")
plt.legend(title="Vehicle Type",loc="upper right")
plt.xticks(range(24))
plt.show()
```

```python
plt.figure(figsize=(10,6))
sns.countplot(data=df, x="Day of the week", hue="Traffic Situation", palette="Set2")
plt.xlabel("Day of the week")
plt.ylabel("Count")
plt.title("Traffic Situation by Day of the Week")
plt.legend(title="Traffic Situation")
plt.xticks(rotation=45)
plt.show()
```
